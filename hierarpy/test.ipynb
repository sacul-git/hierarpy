```python
import pandas as pd
import hierarpy as hp
import tabulate
%matplotlib inline
```

```python
# load an example dataframe
df = pd.read_csv('test_dfs/df1.csv')

print(tabulate.tabulate(df.head(), tablefmt="pipe", headers="keys"))

# transform interactions into matrix
mat = hp.matrix_from_dataframe(df, Winner = 'winner', Loser = 'loser')

print(tabulate.tabulate(mat, tablefmt='pipe', headers='keys'))
mat
```

```python
# david's ranks
david = hp.david_ranks(mat)
print(tabulate.tabulate(david, tablefmt='pipe', headers='keys'))
```

```python
# Adagio:
nodes, edges = hp.run_ADAGIO(mat, preprocess_data = False ,plot=True)
ADAGIO = hp.rank_from_graph(nodes, edges, 'top_down')
print(tabulate.tabulate(ADAGIO, tablefmt='pipe', headers='keys'))
```

```python
# Elo scores:
Elo = hp.elo_ranks(df, Winner='winner', Loser='loser',k=100)
Elo.sort_values('ind')
```
